### 1- Retrieve an arbitrary Wikipedia page of "Python" and create a list of links on that page

```python
url = "https://en.wikipedia.org/wiki/Python"
```

```python
response = requests.get(url)
response.status_code # 200 status code means OK!
```

```python
soup = BeautifulSoup(response.content, "html.parser")
soup
```

```python
links = soup.select(".mw-parser-output > ul > li > a")

# Checking the output
for i in range(len(links)):
    print(links[i])    
```

```python
urls = []

base = "https://en.wikipedia.org"

for link in links:
    urls.append(base + link['href'])

urls
```

```python
# another solution
l = ["https://en.wikipedia.org" + i.get('href') for i in soup.select(".mw-parser-output > ul > li > a")]
l
```

#### 4- Display the 20 latest earthquakes info (date, time, latitude, longitude and region name) by the EMSC as a pandas dataframe: 

```python
url = 'https://www.emsc-csem.org/Earthquake/'
```

```python
response = requests.get(url)
response.status_code # 200 status code means OK!
```

```python
soup = BeautifulSoup(response.content, "html.parser")
soup
```

```python
time=[i.get_text().replace("\xa0"," ") for i in soup.select("td.tabev6 > b > a")]
time
```

```python
lat=[i.get_text().replace("\xa0"," ") for i in soup.select("td:nth-child(5)")]
```

```python
latd=[i.get_text() for i in soup.select("td:nth-child(6)")]
```

```python
latitude=list(map(str.__add__, lat, latd))
latitude
```

```python
long=[i.get_text().replace("\xa0"," ") for i in soup.select("td:nth-child(7)")]
```

```python
longd=[i.get_text().replace("\xa0"," ") for i in soup.select("td:nth-child(8)")]
```

```python
longitude=list(map(''.join, zip(long, longd)))
longitude
```

```python
region_name=[i.get_text().replace("\xa0"," ") for i in soup.select("td.tb_region")]
region_name
```

```python
data=pd.DataFrame({'Date & Time':time, 'latitude':latitude, 'longitude':longitude,'Region Name':region_name})
data
```
